```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

image_path1 = '/content/drive/My Drive/Kmeans_quelccaya_2023/kmeans_2022-06-13.tif'
image_path2 = '/content/drive/My Drive/Kmeans_quelccaya_2023/kmeans_2022-10-11.tif'

image1 = Image.open(image_path1)
image2 = Image.open(image_path2)

image_array1 = np.array(image1)
image_array2 = np.array(image2)

rgb_palette = [
    (1, 1, 1),  # Fundo
    (1, 1, 1),  # Gelo: Branco
    (0.502, 0.502, 0.502),  # Neve: Cinza
    (0.988, 0.553, 0.349),  # Rocha: Laranja
    (0.0, 0.0, 1.0),  # Água: Azul
    (0, 1, 0),  # Crostas: Verde claro
    (0.0, 0.502, 0.0),  # Vegetação: Verde Escuro
]
names = ['Fundo', 'Gelo', 'Neve', 'Rocha', 'Água', 'Crostas', 'Vegetação']

def apply_palette(image_array, rgb_palette):
    colored_image = np.zeros((image_array.shape[0], image_array.shape[1], 3), dtype=np.uint8)
    for i in range(1, len(rgb_palette)):
        colored_image[image_array == i] = np.array(rgb_palette[i]) * 255
    return colored_image

colored_image1 = apply_palette(image_array1, rgb_palette)
colored_image2 = apply_palette(image_array2, rgb_palette)

def print_classes(image_array, names, rgb_palette):
    print("Dados de cada classe e a cor associada:")
    for value in np.unique(image_array):
        if value != 0:
            if value < len(names):
                print(f"Classe: {names[value]} -> Valor: {value} -> Cor: {rgb_palette[value]}")

print("Imagem 1:")
print_classes(image_array1, names, rgb_palette)

print("\nImagem 2:")
print_classes(image_array2, names, rgb_palette)

legend_labels = [
    mpatches.Patch(color=(0.502, 0.502, 0.502), label="Neve (Cinza)"),
    mpatches.Patch(color=(0.988, 0.553, 0.349), label="Rocha (Laranja)"),
    mpatches.Patch(color=(0, 0, 1), label="Água (Azul)"),
    mpatches.Patch(color=(0, 1, 0), label="Crostas (Verde Claro)"),
    mpatches.Patch(color=(0.0, 0.502, 0.0), label="Vegetação (Verde Escuro)"),
]

fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].imshow(colored_image1)
axes[0].set_title("Imagem 1: Junho de 2022")
axes[0].axis('off')

axes[1].imshow(colored_image2)
axes[1].set_title("Imagem 2: Outubro de 2022")
axes[1].axis('off')

plt.legend(handles=legend_labels, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from PIL import Image

image_path1 = '/content/drive/My Drive/Kmeans_quelccaya_2023/kmeans_2022-06-13.tif'
image_path2 = '/content/drive/My Drive/Kmeans_quelccaya_2023/kmeans_2022-10-11.tif'

image1 = Image.open(image_path1)
image2 = Image.open(image_path2)

image_array1 = np.array(image1)
image_array2 = np.array(image2)

names = ['Fundo', 'Gelo', 'Neve', 'Rocha', 'Água', 'Crostas', 'Vegetação']
rgb_palette = [
    (1, 1, 1),
    (0.8, 0.8, 0.8),
    (0.5, 0.5, 0.5),
    (1, 0.56, 0.35),
    (0, 0, 1),
    (0, 1, 0),
    (0.1, 0.3, 0.1),
]

def count_classes(image_array):
    class_counts = {}
    for value in np.unique(image_array):
        if value != 0:
            class_counts[value] = np.sum(image_array == value)
    return class_counts

class_counts_image1 = count_classes(image_array1)
class_counts_image2 = count_classes(image_array2)

flows = []
labels = []
colors = []
sources = []
targets = []

for i in range(1, len(names)):
    for j in range(1, len(names)):
        count1 = np.sum((image_array1 == i) & (image_array2 == j))

        if count1 > 0:
            flows.append(count1)
            sources.append(i - 1)
            targets.append(j - 1 + len(names) - 1)
            labels.append(f'{names[i]} -> {names[j]}')
            rgb_color = rgb_palette[i]
            color_str = f'rgb({int(rgb_color[0] * 255)},{int(rgb_color[1] * 255)},{int(rgb_color[2] * 255)})'
            colors.append(color_str)

fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=names[1:] + names[1:],
        color=['rgba(255, 255, 255, 0)'] * len(names[1:]) + ['rgba(255, 255, 255, 0)'] * len(names[1:]),
        x=[0.1] * (len(names) - 1) + [0.9] * (len(names) - 1),
        y=[i * 0.2 for i in range(len(names) - 1)] + [i * 0.2 + 0.5 for i in range(len(names) - 1)],
    ),
    link=dict(
        source=sources,
        target=targets,
        value=flows,
        color=colors,
        label=labels,
    )
))

fig.update_layout(
    title_text="Diagrama de Sankey - Junho 2022 para outubro de 2022",
    font_size=13,
    height=800,
    width=1200,
    xaxis=dict(showgrid=False, zeroline=False),
    yaxis=dict(showgrid=False, zeroline=False),
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from PIL import Image


image_path1 = '/content/drive/My Drive/Kmeans_quelccaya_2023/kmeans_2022-06-13.tif'
image_path2 = '/content/drive/My Drive/Kmeans_quelccaya_2023/kmeans_2022-10-11.tif'


image1 = Image.open(image_path1)
image2 = Image.open(image_path2)

image_array1 = np.array(image1)
image_array2 = np.array(image2)

names = ['Fundo', 'Gelo', 'Neve', 'Rocha', 'Água', 'Crostas', 'Vegetação']
rgb_palette = [
    (1, 1, 1),  # Fundo (valor 0): Branco
    (0.8, 0.8, 0.8),  # Gelo: Cinza Claro
    (0.5, 0.5, 0.5),  # Neve: Cinza
    (1, 0.56, 0.35),  # Rocha: Laranja
    (0, 0, 1),  # Água: Azul
    (0, 1, 0),  # Crostas: Verde claro
    (0.1, 0.3, 0.1),  # Rocha: Laranja
]


def count_classes(image_array):
    class_counts = {}
    for value in np.unique(image_array):
        if value != 0:  # Ignorar o fundo (valor 0)
            class_counts[value] = np.sum(image_array == value)
    return class_counts

# Contar os pixels de cada classe nas imagens
class_counts_image1 = count_classes(image_array1)
class_counts_image2 = count_classes(image_array2)

# Preparar os dados para o diagrama de Sankey
flows = []  # Lista de fluxos
labels = []  # Lista de rótulos para os fluxos
colors = []  # Lista de cores para os fluxos
sources = []  # Índices das fontes (classe da imagem 1)
targets = []  # Índices dos alvos (classe da imagem 2)

# definir o limiar mínimo
threshold = 4000


for i in range(1, len(names)):
    for j in range(1, len(names)):
        count1 = np.sum((image_array1 == i) & (image_array2 == j))

        if count1 >= threshold:
            flows.append(count1)
            sources.append(i - 1)
            targets.append(j - 1 + len(names) - 1)
            labels.append(f'{names[i]} -> {names[j]}')

            rgb_color = rgb_palette[i]
            color_str = f'rgba({int(rgb_color[0] * 255)},{int(rgb_color[1] * 255)},{int(rgb_color[2] * 255)},0.6)'
            colors.append(color_str)


fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=names[1:] + names[1:],
        color=[f'rgb({int(rgb_palette[i][0] * 255)},{int(rgb_palette[i][1] * 255)},{int(rgb_palette[i][2] * 255)})' for i in range(1, len(names))] + \
              [f'rgb({int(rgb_palette[i][0] * 255)},{int(rgb_palette[i][1] * 255)},{int(rgb_palette[i][2] * 255)})' for i in range(1, len(names))],
        y=[i * 0.2 for i in range(len(names) - 1)] + [i * 0.2 + 0.5 for i in range(len(names) - 1)],
    ),
    link=dict(
        source=sources,  # Fonte dos fluxos
        target=targets,
        value=flows,
        color=colors,
        label=labels,
    )
))


fig.update_layout(
    title_text="Diagrama de Sankey - Junho 2022 para outubro de 2022",
    font_size=13,
    height=800,
    width=1200,
    xaxis=dict(showgrid=False, zeroline=False),
    yaxis=dict(showgrid=False, zeroline=False),
)


fig.show()
```
